```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# you can find references for different types of analyses under each section.

# Load the dataset
ad_sales = pd.read_csv('https://raw.githubusercontent.com/utjimmyx/regression/master/advertising.csv')

# Scatter plot: Sales vs. TV advertising budget
plt.figure(figsize=(8, 6))
sns.scatterplot(x='TV', y='sales', data=ad_sales)
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('Sales vs. TV Advertising Budget')
plt.show()
# Ref: https://seaborn.pydata.org/generated/seaborn.scatterplot.html

# Histogram of TV advertising budgets
plt.figure(figsize=(8, 6))
sns.histplot(ad_sales['TV'], bins=range(0, int(ad_sales['TV'].max()) + 5, 5), color='gray')
plt.xlabel('TV Advertising Budget')
plt.title('Distribution of TV Advertising Budgets')
plt.show()

# Bar plot of TV advertising budgets
plt.figure(figsize=(8, 6))
sns.countplot(x='TV', data=ad_sales, color='gray')
plt.xlabel('TV Advertising Budget')
plt.title('Count of TV Advertising Budgets')
plt.show()
#Ref: https://www.geeksforgeeks.org/bar-plot-in-matplotlib/

# Horizontal bar plot of TV advertising budgets
plt.figure(figsize=(8, 6))
sns.countplot(y='TV', data=ad_sales, color='gray')
plt.ylabel('TV Advertising Budget')
plt.title('Count of TV Advertising Budgets')
plt.show()

# Scatter plot with regression line: Sales vs. TV advertising budget
plt.figure(figsize=(8, 6))
sns.regplot(x='TV', y='sales', data=ad_sales, scatter_kws={'s': 50}, line_kws={'color': 'blue'})
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('Sales vs. TV Advertising Budget with Regression Line')
plt.show()
## reference: https://www.geeksforgeeks.org/python-seaborn-regplot-method/
```
